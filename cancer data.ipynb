```python
import pandas as pd 
import numpy as np
```

```python
from sklearn.datasets import  load_breast_cancer
```

```python
cancer_data = load_breast_cancer()
```

```python
df = pd.DataFrame(data = cancer_data.data, columns = cancer_data.feature_names)
```

```python
df.head()
```

```python
df['target']= cancer_data.target
```

```python
df.head()
```

```python
import sweetviz as sv
```

```python
report = sv.analyze(df)
```

```python
report.show_html('dATA.HTML')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer_data.data, cancer_data.target, test_size=0.25)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
pred = lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
accuracy_score(y_pred = pred, y_true= y_test)
```

```python
confusion_matrix(y_pred=pred, y_true = y_test)
```
